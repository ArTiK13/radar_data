```python
import numpy as np
import scipy
import json
import matplotlib.pyplot as plt
import pandas as pd
```

```python
radar_df = pd.read_csv("../data/processed data/radar_data_7.csv")
lidar_df = pd.read_csv("../data/processed data/lidar_data_7.csv")
```

```python
plt.figure(figsize=(15, 10))

dff = list(
    radar_df[radar_df["AbsoluteRadialVelocity"] > 1][["X, (m)", "Y, (m)"]].itertuples(
        index=False, name=None
    )
)

a = scipy.spatial.distance_matrix(dff, dff)
np.fill_diagonal(a, 100)

t = [np.partition(v, 3)[3] for v in a]
x = []
y = []
for i, k in enumerate(dff):
    if t[i] < 1:
        x.append(k[0])
        y.append(k[1])
plt.scatter(x, y, c="lime", s=1, label="3 соседа")
```
